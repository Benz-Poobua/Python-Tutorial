````python
x1 = np.linspace(-5,5,100)
plt.plot(x1, x1**2, 'ko', label = 'quadratic')
plt.plot(x1, x1**3, 'r*', label = 'cubic')
plt.plot(x1, x1**2 + 50*np.cos(x1+30), 'b-', label = 'polynomial')
#plt.semilogx(x,y)
plt.title(f'Plot of Various Polynomials from {x1[0]} to {x1[-1]}')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc = 2) #position from 0 to 10
plt.xlim(-6,6)
plt.ylim(-10,10)
plt.grid()
plt.show()
````

````python
fig, ax = plt.subplots(2,1)

x2 = np.linspace(1,10,50)
y2 = np.sin(x2)
z2 = np.cos(x2)

ax[0].plot(x2,y2, 'r:', label='sin(x2)')
ax[0].set_title('Sine with Differing Smoothness')
ax[0].set_xlabel('x (radians)')
ax[0].set_ylabel('Sine of x')
ax[0].legend()

ax[1].plot(x2, z2, 'r:', label='cos(x2)')
ax[1].set_xlabel('x (radians)')
ax[1].set_ylabel('Cosine of x')
ax[1].legend()

plt.tight_layout() 
plt.show()
````

````python
import requests, io
````

````python
url = 'https://ncedc.org/ftp/pub/doc/NC.info/NC.channel.summary.day'
# this gets the file linked in the URL page and convert it to a string
s = requests.get(url).content 

# this will convert the string, decode it , and make it a table
data = pd.read_csv(io.StringIO(s.decode('utf-8')), header=None, skiprows=2, sep='\s+', usecols=list(range(0, 13)))
# because columns/keys were not assigned, assign them now
data.columns = ['station', 'network', 'channel', 'location', 'rate', 'start_time', 'end_time', 'latitude', 'longitude', 'elevation', 'depth', 'dip', 'azimuth']
````

````python
data.head()
````

````python
data.iloc[0] # we can output the first element of the DataFrame
````

````python
data.loc[data['channel']=='EHZ']
````

````python
data.iloc[:, 0] # r, c = all rows + first col
````

````python
# Select two stations, use the typical "OR" |
data.loc[(data.station=='KCPB') | (data.station=='KHBB')]

# OR |
# AND &
````

````python
data.loc[data.station=='KCPB']
````

````python
# check for NaN (Not a Number) values in the entire DataFrame
nan_check = data.isna()
print(nan_check)
````

````python
nan_counts = data.isna().sum()
print(nan_counts)
````

````python
# The DataFrame may have bad values. A typical data cleaning involves removing Nan and Zeros
# drop rows with any NaN values
# update hte original DataFrame directly without needing to assign it to a new variable
data.dropna(inplace=True) 
data=data[data.longitude!=0]
````

````python
South_Beach_2009 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/South_Beach/South_Beach_2009.csv')
South_Beach_2010 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/South_Beach/South_Beach_2010.csv')
South_Beach_2011 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/South_Beach/South_Beach_2011.csv')
````

````python
South_Beach_2009_2011 = pd.concat([South_Beach_2009, South_Beach_2010, South_Beach_2011], ignore_index=True)
South_Beach_2009_2011['Datetime'] = pd.to_datetime(South_Beach_2009_2011['Date'] + ' ' + South_Beach_2009_2011['Time (GMT)'])
````

````python
Astoria_2006 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/Astoria/Astoria_2006.csv')
Astoria_2007 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/Astoria/Astoria_2007.csv')
Astoria_2008 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/Astoria/Astoria_2008.csv')
Astoria_2009 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/Astoria/Astoria_2009.csv')
Astoria_2010 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/Astoria/Astoria_2010.csv')
Astoria_2011 = pd.read_csv('https://raw.githubusercontent.com/Benz-Poobua/Undergraduate-Research-2/main/NOAA_WL/OR/Astoria/Astoria_2011.csv')
````

````python
Astoria_2006_2011 = pd.concat([Astoria_2006, Astoria_2007, Astoria_2008, Astoria_2009, Astoria_2010, Astoria_2011], ignore_index=True)
Astoria_2006_2011['Datetime'] = pd.to_datetime(Astoria_2006_2011['Date'] + ' ' + Astoria_2006_2011['Time (GMT)'])
````

````python
fig,ax=plt.subplots(2,1,figsize=(11,8),sharex=True)
fig.suptitle('Northern Oregon from 2006 to 2011')
ax[0].plot(South_Beach_2009_2011['Datetime'], South_Beach_2009_2011['Verified (ft)'], '-.b');ax[0].grid(True);ax[0].set_ylabel('South Beach [ft]') 
ax[1].plot(Astoria_2006_2011['Datetime'], Astoria_2006_2011['Verified (ft)'], '-.b');ax[1].grid(True);ax[1].set_ylabel('Astoria [ft]') 
ax[1].set_xlabel('Time (years)')
plt.show()
````
